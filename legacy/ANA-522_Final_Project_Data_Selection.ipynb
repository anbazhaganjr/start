```python
import os
import glob
import pandas as pd
from tabulate import tabulate

# Configuration block
config = {
    "DATA_FOLDER_PATH": r"/Users/anbzhnjr/learning/python_522/final_project/data/historical/",
    #"DATA_FOLDER_PATH": r"/",
    "STOCK_SYMBOLS": {
        'NVIDIA': 'NVDA', 'TESLA': 'TSLA', 'APPLE': 'AAPL', 'MICROSOFT': 'MSFT', 'ALPHABET': 'GOOGL',
        'AMAZON': 'AMZN', 'META': 'META', 'BERKSHIRE': 'BRKB', 'BROADCOM': 'AVGO', 'UNITEDHEALTH': 'UNH'
    }
}

# Access configuration variables
main_folder_path = config.get("DATA_FOLDER_PATH")
stock_name_to_symbol = config.get("STOCK_SYMBOLS")

# Dictionary to store DataFrames for each stock symbol
data_store = {}

# Variable metadata for tabular display
variable_metadata = {
    "Variable Name": ["timestamp", "price", "open", "high", "low", "close", "volume", "vwap", "symbol"],
    "Description": [
        "The date and time of each observation, allowing for time-based analysis in intraday trading",
        "The trading price of the stock at the specific timestamp",
        "The opening price of the stock at the start of each interval",
        "The highest trading price of the stock during each interval",
        "The lowest trading price of the stock during each interval",
        "The closing price of the stock at the end of each interval",
        "The total volume of shares traded during each interval, indicating trading activity",
        "The Volume Weighted Average Price, reflecting the average price adjusted by trading volume",
        "The stock ticker symbol, identifying the specific stock for each observation"
    ],
    "Type": [
        "Categorical (DateTime)", "Continuous", "Continuous", "Continuous", 
        "Continuous", "Continuous", "Continuous", "Continuous", "Categorical"
    ]
}
variable_info_df = pd.DataFrame(variable_metadata)

# Function to dynamically generate a summary with data quality checks
def generate_summary(df, stock_symbol):
    numerical_stats = df.describe()
    total_rows, total_columns = df.shape
    missing_values = df.isnull().sum()
    missing_info = missing_values[missing_values > 0]
    data_issues = []

    # Check for missing values
    if not missing_info.empty:
        issue = f"Missing values detected in the following columns:\n{missing_info.to_string()}"
        data_issues.append(issue)

    # Check for duplicate rows
    duplicate_count = df.duplicated().sum()
    if duplicate_count > 0:
        data_issues.append(f"{duplicate_count} duplicate rows found.")

    # Generate the summary text dynamically
    summary = f"""

The dataset for "{stock_symbol}" contains {total_rows} rows and {total_columns} columns. Key metrics include:
- **Mean Price**: {numerical_stats.loc['mean', 'price']:.2f}
- **Max Price**: {numerical_stats.loc['max', 'price']:.2f}
- **Min Price**: {numerical_stats.loc['min', 'price']:.2f}
- **Average Volume**: {numerical_stats.loc['mean', 'volume']:.2f}

---

"""

    if data_issues:
        summary += "\nThe following data quality issues were identified:\n"
        summary += "\n".join(f"- {issue}" for issue in data_issues)
    else:
        summary += "The dataset is clean, with no significant issues detected."

    return summary

# Data Loading
print("\nLoading data for all stock symbols...\n")
for stock_symbol in stock_name_to_symbol.values():
    stock_folder_path = os.path.join(main_folder_path, stock_symbol)
    if not os.path.exists(stock_folder_path):
        print(f"Folder for stock symbol '{stock_symbol}' not found. Skipping...")
        continue

    # Find all JSON and CSV files
    all_files = glob.glob(os.path.join(stock_folder_path, "*.json")) + glob.glob(os.path.join(stock_folder_path, "*.csv"))
    stock_data = []

    for file in all_files:
        try:
            if file.endswith(".json"):
                df = pd.read_json(file)
            elif file.endswith(".csv"):
                df = pd.read_csv(file)
            df['symbol'] = stock_symbol
            stock_data.append(df)
        except Exception as e:
            print(f"Error processing file {file}: {e}")
    
    if stock_data:
        data_store[stock_symbol] = pd.concat(stock_data, ignore_index=True)

print("\nData loading complete.")

# User Interaction and Analysis
user_input = input("\nEnter one or more stock names (comma-separated) to analyze, or type 'exit' to quit: ").strip()
if user_input.lower() == "exit":
    print("Exiting. Goodbye!")
else:
    # Process user input
    input_names = [name.strip().upper() for name in user_input.split(",")]
    input_symbols = [stock_name_to_symbol.get(name) for name in input_names]
    valid_symbols = [symbol for symbol in input_symbols if symbol and symbol in data_store]
    invalid_names = [name for name, symbol in zip(input_names, input_symbols) if not symbol]

    # Analyze valid symbols
    for stock_symbol in valid_symbols:
        df = data_store[stock_symbol]
        print(f"\n**Displaying data for stock symbol: {stock_symbol}**\n")

        # Additional Dataset Information
        print("\n**Dataset Information**")
        df.info()

        print("\n**Dataset Shape**")
        print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

        print("\n**Column Names**")
        print(df.columns.tolist())

        print("\n**Data Types**")
        print(df.dtypes)

        print("\n**Variable Information**")
        print(tabulate(variable_info_df, headers='keys', tablefmt='grid'))

        # Basic Dataset Examination
        print("\n**Basic Dataset Examination**")

        # First and Last 5 Records
        print("\n**First 5 Records:**")
        print(tabulate(df.head(), headers="keys", tablefmt="fancy_grid"))
        print("\n**Last 5 Records:**")
        print(tabulate(df.tail(), headers="keys", tablefmt="fancy_grid"))

        # Summary Statistics
        print("\n**Summary Statistics for Numerical Variables (Preview):**")
        print(tabulate(df.describe().head(), headers="keys", tablefmt="fancy_grid"))

        # Generate and display dynamic summary
        summary = generate_summary(df, stock_symbol)
        print(summary)

    # Notify about invalid inputs
    if invalid_names:
        print(f"\nInvalid stock names entered: {', '.join(invalid_names)}")
```
